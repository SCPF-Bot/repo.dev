# MLBB Hero Portrait Classifier — Training Pipeline

**Produces:** `mlbb_hero_classifier.tflite` + `hero_classifier_labels.txt`  
Drop both files into `app/src/main/assets/` and ship.

**Runtime:** ~30–60 min on Colab T4 GPU (free tier).

**Workflow:**
1. Upload `default_heroes.json` from your repo (`app/src/main/res/raw/`)
2. Run all cells top-to-bottom
3. Download `mlbb_hero_classifier.tflite` **and** `hero_classifier_labels.txt` at the end
4. Place both files in `app/src/main/assets/` — the app requires both

**Accuracy expectations:**
- CDN portraits only (no screenshots): ~75–85 % top-1
- CDN + your own draft screen crops:   ~92–97 % top-1
- See Cell 4 for how to add screenshot crops

**Android contract (do not change without updating `HeroClassifier.kt`):**
- Input tensor  : `[1, 224, 224, 3]` float32, pixels normalised to [−1, 1] via `(px / 127.5) − 1`
- Output tensor : `[1, N]` float32 softmax probabilities, one entry per hero class
- Label order   : heroes sorted ascending by `id` field — this order defines `heroIds` in `PortraitMatcher`

## Cell 3 — Download CDN portrait images

Downloads one portrait per hero from the `imageUrl` field.
Skips already-downloaded files so this cell is safe to re-run.

````python
import requests, time
from pathlib import Path
from tqdm import tqdm

CDN_DIR = Path("hero_data/cdn")
CDN_DIR.mkdir(parents=True, exist_ok=True)

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Linux; Android 10) AppleWebKit/537.36 "
        "Chrome/120.0.0.0 Mobile Safari/537.36"
    )
}

failed = []
for hero in tqdm(heroes, desc="Downloading CDN portraits"):
    dest = CDN_DIR / f"{hero['id']}.png"
    if dest.exists():
        continue
    try:
        r = requests.get(hero["imageUrl"], headers=HEADERS, timeout=15)
        r.raise_for_status()
        dest.write_bytes(r.content)
        time.sleep(0.15)   # be polite to the CDN
    except Exception as e:
        failed.append((hero["name"], str(e)))

if failed:
    print(f"⚠  {len(failed)} downloads failed:")
    for name, err in failed:
        print(f"   {name}: {err}")
else:
    downloaded = len(list(CDN_DIR.glob("*.png")))
    print(f"✓  {downloaded}/{NUM_CLASSES} portraits saved to {CDN_DIR}")
````

## Cell 5 — Build the dataset

For each hero we collect all available images (CDN + screenshots), then apply
heavy augmentation to synthesise a larger training set.

Augmentations chosen specifically for draft-screen conditions:
- JPEG quality drop (simulates MediaProjection compression artefacts)
- Brightness / contrast shift (simulates adaptive-brightness displays)
- Random crop + zoom (simulates slot-region misalignment)
- Horizontal flip (some heroes face left in certain skins)
- Gaussian noise (simulates low-bitrate capture)

````python
import tensorflow as tf
import numpy as np
from PIL import Image, ImageEnhance, ImageFilter
import io as _io
import random

tf.get_logger().setLevel("ERROR")

IMG_SIZE        = 224    # must match HeroClassifier.INPUT_SIZE (HeroClassifier.kt)
AUGMENT_PER_SRC = 40    # synthetic copies per source image
BATCH_SIZE      = 32
VALIDATION_SPLIT = 0.15  # 15 % of images held out for validation

# ── Augmentation function (PIL-based for JPEG sim) ───────────────────────────

def augment_pil(img: Image.Image) -> Image.Image:
    """Apply random augmentations to a PIL image."""
    img = img.convert("RGB")

    # Random horizontal flip
    if random.random() < 0.5:
        img = img.transpose(Image.FLIP_LEFT_RIGHT)

    # Brightness
    factor = random.uniform(0.7, 1.4)
    img = ImageEnhance.Brightness(img).enhance(factor)

    # Contrast
    factor = random.uniform(0.8, 1.3)
    img = ImageEnhance.Contrast(img).enhance(factor)

    # Saturation (handles skin colour palette shifts)
    factor = random.uniform(0.7, 1.4)
    img = ImageEnhance.Color(img).enhance(factor)

    # Sharpness / slight blur (simulates MediaProjection blur)
    if random.random() < 0.3:
        img = img.filter(ImageFilter.GaussianBlur(radius=random.uniform(0.5, 1.5)))

    # Random rotation ±12°
    angle = random.uniform(-12, 12)
    img = img.rotate(angle, resample=Image.BILINEAR, fillcolor=(0, 0, 0))

    # Random crop (crop 80–100 % of the image, then resize back)
    w, h = img.size
    crop_frac = random.uniform(0.82, 1.0)
    cw, ch = int(w * crop_frac), int(h * crop_frac)
    left   = random.randint(0, w - cw)
    top    = random.randint(0, h - ch)
    img    = img.crop((left, top, left + cw, top + ch)).resize((w, h), Image.BILINEAR)

    # JPEG quality simulation (the big one for MediaProjection accuracy)
    quality = random.randint(55, 92)
    buf = _io.BytesIO()
    img.save(buf, format="JPEG", quality=quality)
    buf.seek(0)
    img = Image.open(buf).copy()

    # Gaussian noise
    arr = np.array(img, dtype=np.float32)
    noise = np.random.normal(0, random.uniform(0, 8), arr.shape)
    arr = np.clip(arr + noise, 0, 255).astype(np.uint8)
    img = Image.fromarray(arr)

    return img

# ── Build file lists ──────────────────────────────────────────────────────────

DATASET_DIR = Path("hero_data/dataset")
DATASET_DIR.mkdir(parents=True, exist_ok=True)

all_paths, all_labels = [], []

print("Building augmented dataset …")
for label_idx, hero in enumerate(tqdm(heroes)):
    hero_id   = hero["id"]
    class_dir = DATASET_DIR / str(hero_id)
    class_dir.mkdir(exist_ok=True)

    # Collect source images
    sources = []
    cdn_path = CDN_DIR / f"{hero_id}.png"
    if cdn_path.exists():
        sources.append(cdn_path)
    if USE_SCREENSHOTS:
        for ext in ("*.jpg", "*.jpeg", "*.png"):
            sources.extend((SCREENSHOTS_DIR / str(hero_id)).glob(ext))

    if not sources:
        print(f"  ⚠  No images for hero {hero['name']} (ID {hero_id}) — skipping")
        continue

    # Write augmented copies
    aug_count = 0
    for src_path in sources:
        try:
            src_img = Image.open(src_path).convert("RGB").resize((IMG_SIZE, IMG_SIZE), Image.BILINEAR)
        except Exception:
            continue

        # Always keep the original (no augmentation)
        orig_dest = class_dir / f"orig_{src_path.stem}.jpg"
        if not orig_dest.exists():
            src_img.save(orig_dest, quality=95)
        all_paths.append(str(orig_dest))
        all_labels.append(label_idx)

        # Augmented copies
        for i in range(AUGMENT_PER_SRC):
            aug_dest = class_dir / f"aug_{src_path.stem}_{i:03d}.jpg"
            if not aug_dest.exists():
                augment_pil(src_img).save(aug_dest, quality=90)
            all_paths.append(str(aug_dest))
            all_labels.append(label_idx)
            aug_count += 1

print(f"\n✓  Dataset: {len(all_paths)} images across {NUM_CLASSES} classes")
print(f"   ~{len(all_paths) // NUM_CLASSES} images/hero on average")
````

## Cell 6 — Create train / validation splits

````python
from sklearn.model_selection import train_test_split

paths_arr  = np.array(all_paths)
labels_arr = np.array(all_labels)

train_paths, val_paths, train_labels, val_labels = train_test_split(
    paths_arr, labels_arr,
    test_size    = VALIDATION_SPLIT,
    random_state = 42,
    stratify     = labels_arr,   # keep class proportions balanced
)

print(f"Train : {len(train_paths):>6,} images")
print(f"Val   : {len(val_paths):>6,} images")
print(f"Ratio : {len(val_paths)/len(train_paths):.1%} val")

# ── TF Dataset pipeline ───────────────────────────────────────────────────────

def load_and_preprocess(path, label):
    raw  = tf.io.read_file(path)
    img  = tf.image.decode_jpeg(raw, channels=3)
    img  = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    # NOTE: tf.keras.applications.mobilenet_v3.preprocess_input is a NO-OP —
    # MobileNetV3's scaling normally happens inside the model via a Rescaling
    # layer, gated by include_preprocessing. We set include_preprocessing=False
    # in Cell 7, so we must do the [-1, 1] scaling ourselves here.
    # This formula MUST match HeroClassifier.preprocessBitmap():
    #   input[y][x][c] = (channel_value / 127.5f) - 1f
    img  = (tf.cast(img, tf.float32) - 127.5) / 127.5
    return img, label

def make_dataset(paths, labels, shuffle=False):
    ds = tf.data.Dataset.from_tensor_slices((paths, labels))
    if shuffle:
        ds = ds.shuffle(buffer_size=len(paths), seed=42, reshuffle_each_iteration=True)
    ds = (ds
          .map(load_and_preprocess, num_parallel_calls=tf.data.AUTOTUNE)
          .batch(BATCH_SIZE)
          .prefetch(tf.data.AUTOTUNE))
    return ds

train_ds = make_dataset(train_paths, train_labels, shuffle=True)
val_ds   = make_dataset(val_paths,   val_labels,   shuffle=False)

print("✓  tf.data pipelines ready")
````

## Cell 7 — Build model (MobileNetV3-Small + classification head)

MobileNetV3-Small is chosen because:
- Small footprint (2.5 MB float32, ~1.2 MB float16 TFLite)
- Fast inference on CPU (~5–15 ms, fits in FrameProcessor budget)
- Pretrained ImageNet weights transfer well to portrait classification

Training is two-phase:
1. **Head-only** (12 epochs): base frozen, only the Dense head trains.
2. **Fine-tune** (25 epochs): last 30 layers unfrozen, lower LR.

````python
from tensorflow.keras import layers, Model
import tensorflow.keras.applications as apps

# Phase 1 setup
base = apps.MobileNetV3Small(
    input_shape           = (IMG_SIZE, IMG_SIZE, 3),
    include_top           = False,
    weights               = "imagenet",
    include_preprocessing = False,  # we do [-1,1] ourselves in load_and_preprocess
)
base.trainable = False

inputs  = tf.keras.Input(shape=(IMG_SIZE, IMG_SIZE, 3))
x       = base(inputs, training=False)
x       = layers.GlobalAveragePooling2D()(x)
x       = layers.BatchNormalization()(x)
x       = layers.Dropout(0.35)(x)
outputs = layers.Dense(NUM_CLASSES, activation="softmax")(x)

model = Model(inputs, outputs, name="mlbb_hero_classifier")
model.summary(line_length=80)
````

## Cell 8 — Phase 1 training: head only

````python
import os

CKPT_DIR = Path("checkpoints")
CKPT_DIR.mkdir(exist_ok=True)

callbacks_phase1 = [
    tf.keras.callbacks.ModelCheckpoint(
        str(CKPT_DIR / "phase1_best.keras"),
        monitor="val_accuracy", save_best_only=True, verbose=1
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy", patience=4, restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.5, patience=2, min_lr=1e-6, verbose=1
    ),
]

model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=1e-3),
    loss      = "sparse_categorical_crossentropy",
    metrics   = ["accuracy"],
)

print("Phase 1: training head only …")
history1 = model.fit(
    train_ds,
    validation_data = val_ds,
    epochs          = 12,
    callbacks       = callbacks_phase1,
)

phase1_acc = max(history1.history["val_accuracy"])
print(f"\n✓  Phase 1 best val accuracy: {phase1_acc:.1%}")
````

## Cell 9 — Phase 2 fine-tuning: unfreeze last 30 layers

````python
# Unfreeze the last 30 layers of the base model
base.trainable = True
for layer in base.layers[:-30]:
    layer.trainable = False

trainable_count = sum(1 for l in base.layers if l.trainable)
print(f"Trainable base layers: {trainable_count}/{len(base.layers)}")

callbacks_phase2 = [
    tf.keras.callbacks.ModelCheckpoint(
        str(CKPT_DIR / "phase2_best.keras"),
        monitor="val_accuracy", save_best_only=True, verbose=1
    ),
    tf.keras.callbacks.EarlyStopping(
        monitor="val_accuracy", patience=6, restore_best_weights=True
    ),
    tf.keras.callbacks.ReduceLROnPlateau(
        monitor="val_loss", factor=0.4, patience=3, min_lr=1e-7, verbose=1
    ),
]

# Much lower LR for fine-tuning to avoid destroying pretrained weights
model.compile(
    optimizer = tf.keras.optimizers.Adam(learning_rate=5e-5),
    loss      = "sparse_categorical_crossentropy",
    metrics   = ["accuracy"],
)

print("\nPhase 2: fine-tuning last 30 layers …")
history2 = model.fit(
    train_ds,
    validation_data = val_ds,
    epochs          = 25,
    callbacks       = callbacks_phase2,
)

phase2_acc = max(history2.history["val_accuracy"])
print(f"\n✓  Phase 2 best val accuracy: {phase2_acc:.1%}")
````

## Cell 10 — Validation report

````python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix

# Collect predictions on the validation set
print("Running validation inference …")
y_true, y_pred = [], []
for images, labels in val_ds:
    preds = model.predict(images, verbose=0)
    y_true.extend(labels.numpy().tolist())
    y_pred.extend(np.argmax(preds, axis=1).tolist())

y_true = np.array(y_true)
y_pred = np.array(y_pred)

top1_acc = np.mean(y_true == y_pred)
print(f"\nFinal top-1 accuracy on validation set: {top1_acc:.1%}")

if top1_acc < 0.80:
    print("⚠  Accuracy below 80 % — consider adding screenshot crops (Cell 4) and retraining.")
elif top1_acc < 0.92:
    print("⚠  Accuracy < 92 % — usable but screenshot crops will push this higher.")
else:
    print("✓  Accuracy ≥ 92 % — ready for production.")

# Per-hero accuracy (find the worst-performing heroes)
print("\n── Worst 10 heroes by accuracy ──────────────────────────────────")
per_hero_acc = {}
for label_idx, hero in enumerate(heroes):
    mask    = y_true == label_idx
    if mask.sum() == 0:
        continue
    correct = np.sum(y_pred[mask] == label_idx)
    per_hero_acc[hero["name"]] = correct / mask.sum()

for name, acc in sorted(per_hero_acc.items(), key=lambda x: x[1])[:10]:
    print(f"  {name:<20} {acc:.1%}")

# Confusion matrix (top-20 most confused classes for readability)
confused_pairs = []
for i in range(NUM_CLASSES):
    mask = y_true == i
    if mask.sum() == 0:
        continue
    wrong = y_pred[mask]
    for j in np.unique(wrong):
        if j != i:
            confused_pairs.append((HERO_NAMES[i], HERO_NAMES[j], np.sum(wrong == j)))

confused_pairs.sort(key=lambda x: -x[2])
print("\n── Top 10 confusions ────────────────────────────────────────────")
for true_name, pred_name, count in confused_pairs[:10]:
    print(f"  {true_name:<20} mistaken for  {pred_name:<20}  ({count}×)")
````

## Cell 11 — Convert to TFLite (float16 quantisation)

float16 quantisation:
- Cuts model size by ~50 % vs float32
- Compatible with the GPU delegate in `HeroClassifier.kt`
- Negligible accuracy loss (< 0.5 % on validation)

````python
print("Converting to TFLite …")

converter = tf.lite.TFLiteConverter.from_keras_model(model)
converter.optimizations       = [tf.lite.Optimize.DEFAULT]
converter.target_spec.supported_types = [tf.float16]

tflite_model = converter.convert()

OUTPUT_PATH = Path("mlbb_hero_classifier.tflite")
OUTPUT_PATH.write_bytes(tflite_model)

size_mb = len(tflite_model) / 1_048_576
print(f"✓  Saved: {OUTPUT_PATH}  ({size_mb:.2f} MB)")
print(f"   Target: < 3 MB  {'✓' if size_mb < 3 else '⚠ larger than expected'}")
````

## Cell 12 — Validate the TFLite model

Runs the TFLite model on the validation set via the TFLite Interpreter
(not Keras) to confirm the converted model behaves identically.

````python
interpreter = tf.lite.Interpreter(model_path=str(OUTPUT_PATH))
interpreter.allocate_tensors()

input_details  = interpreter.get_input_details()
output_details = interpreter.get_output_details()

print("TFLite input  shape:", input_details[0]["shape"])
print("TFLite output shape:", output_details[0]["shape"])
print(f"Expected output: [1, {NUM_CLASSES}]")

# Spot-check on 200 random validation images
sample_size = min(200, len(val_paths))
indices     = np.random.choice(len(val_paths), sample_size, replace=False)

tflite_correct = 0
for idx in indices:
    raw  = tf.io.read_file(val_paths[idx]).numpy()
    img  = tf.image.decode_jpeg(raw, channels=3)
    img  = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img  = (tf.cast(img, tf.float32) - 127.5) / 127.5
    inp  = tf.expand_dims(img, 0)                                 # [1, 224, 224, 3]

    interpreter.set_tensor(input_details[0]["index"], inp)
    interpreter.invoke()
    probs = interpreter.get_tensor(output_details[0]["index"])[0]  # [NUM_CLASSES]

    tflite_pred = np.argmax(probs)
    if tflite_pred == val_labels[idx]:
        tflite_correct += 1

tflite_acc = tflite_correct / sample_size
print(f"\nTFLite spot-check accuracy ({sample_size} samples): {tflite_acc:.1%}")

delta = abs(top1_acc - tflite_acc)
print(f"Delta vs Keras model: {delta:.1%}  {'✓ within 1%' if delta < 0.01 else '⚠ larger than expected — check quantisation'}")
````

## Cell 13b — Write `hero_classifier_labels.txt` (required by the Android app)

This file is what `HeroClassifier.kt` actually reads at runtime.
Format: one hero ID (integer) per line, in the same ascending-ID order
used during training.

**This file must be copied to `app/src/main/assets/hero_classifier_labels.txt`
every time the model is retrained**, especially when new heroes are added.
`HeroClassifier` validates that the line count equals the model output size
and disables the TFLite path (falling back to pHash) if they differ.

````python
LABELS_TXT_PATH = Path("hero_classifier_labels.txt")

lines = [
    "# MLBB Hero Classifier — output index → heroId mapping",
    "#",
    "# Each non-comment, non-blank line maps one output neuron (by line order,",
    "# 0-indexed) to an MLBB hero ID as stored in HeroEntity.id / Hero.id.",
    "#",
    f"# Model: mlbb_hero_classifier.tflite — {NUM_CLASSES} classes",
    "# Heroes sorted ascending by ID (matches default_heroes.json at training time).",
    "#",
    "# Generated automatically by the training notebook — do not edit by hand.",
    "# Regenerate by rerunning this notebook whenever heroes are added or removed.",
    "#",
    "# Format: one integer (heroId) per line. Comment lines (#) and blank lines",
    "# are skipped by HeroClassifier.kt.",
]
lines += [str(hero["id"]) for hero in heroes]

LABELS_TXT_PATH.write_text("\n".join(lines) + "\n")

# Verify the line count matches NUM_CLASSES
actual_ids = [
    int(l.strip()) for l in LABELS_TXT_PATH.read_text().splitlines()
    if l.strip() and not l.strip().startswith("#")
]
assert len(actual_ids) == NUM_CLASSES, (
    f"BUG: label file has {len(actual_ids)} entries but model has {NUM_CLASSES} classes"
)
assert actual_ids == HERO_IDS, "BUG: heroId order in label file does not match training order"

print(f"✓  hero_classifier_labels.txt saved: {LABELS_TXT_PATH}")
print(f"   {len(actual_ids)} heroId entries — matches model output size ✓")
print(f"   First entry: index 0 → heroId {actual_ids[0]} ({heroes[0]['name']})")
print(f"   Last  entry: index {len(actual_ids)-1} → heroId {actual_ids[-1]} ({heroes[-1]['name']})")
````
